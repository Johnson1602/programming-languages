```python
# 用户输入一个数字
# 打印每一位数字及其重复的次数

num = input('>>> ')
repeat = {}
for c in num:
    if c not in repeat:
        repeat[c] = 1
    else:
        repeat[c] += 1
for c in num:
    print('{}: appeared {} time(s)'.format(c, repeat[c]))
```

```python
# 数字重复统计
# 随机产生100个整数
# 数字的范围[-1000, 1000]
# 升序输出所有不同的数字及其重复的次数

from random import randint

nums = [0] * 100
repeat = {}
for i in range(100):
    num = randint(-1000, 1000)
    nums[i] = num
    if num not in repeat:
        repeat[num] = 1
    else:
        repeat[num] += 1
nums = list(set(nums))
length = len(nums)
for m in range(length // 2):
    maxindex = -m - 1
    minindex = m
    for n in range(m + 1, length - m):
        if nums[n] > nums[maxindex]:
            maxindex = n
        if nums[n] < nums[minindex]:
            minindex = n
    if nums[maxindex] == nums[minindex]:
        break
    if maxindex != m:
        nums[m], nums[maxindex] = nums[maxindex], nums[m]
        if m == minindex:
            minindex = maxindex
    if minindex != -m - 1 and nums[minindex] != nums[-m - 1]:
        nums[-m - 1], nums[minindex] = nums[minindex], nums[-m - 1]
print('There are {} different numbers.'.format(len(nums)))
for j in nums:
    print('{} appeared {} time(s)'.format(j, repeat[j]))
```

```python
# 字符串重复统计
# 字符表'abcdefghijklmnopqrstuvwxyz'
# 随机挑选2个字母组成字符串，共挑选100个
# 降序输出所有不同的字符串及重复的次数

from random import randrange

alphabet = 'abcdefghijklmnopqrstuvwxyz'
length = len(alphabet)
strings = {}
for i in range(100):
    new_string = alphabet[randrange(length)] + alphabet[randrange(length)]
    if new_string not in strings:
        strings[new_string] = 1
    else:
        strings[new_string] += 1
sorted_strings = sorted(strings.keys(), key=str, reverse=True)
print('There are {} different strings.'.format(len(strings)))
for key in sorted_strings:
    print('String \'{}\' appeared {} time(s).'.format(key, strings[key]))
```
