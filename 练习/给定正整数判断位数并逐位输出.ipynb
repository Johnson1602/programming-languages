```python
# 给定一个不超过5位的正整数，判断该数的位数，依次打印出个位、十位、百位、千位、万位的数字
# 法1
a = int(input('>>> '))

if a >= 1000:
    if a >= 10000:
        print(5)
    else:
        print(4)
else:
    if a >= 100:
        print(3)
    elif a >= 10:
        print(2)
    else:
        print(1)

wan = a // 10000
qian = (a - wan * 10000) // 1000
bai = (a - wan * 10000 - qian * 1000) // 100
shi = (a - wan * 10000 - qian * 1000 - bai * 100) // 10
ge = a - wan * 10000 - qian * 1000 - bai * 100 - shi * 10

print(ge, shi, bai, qian, wan)
```

```python
# 给定一个不超过5位的正整数，判断该数的位数，依次打印出个位、十位、百位、千位、万位的数字
# 法2

a = int(input('>>> '))

if a >= 1000:
    if a >= 10000:
        num = 5
    else:
        num = 4
else:
    if a >= 100:
        num = 3
    elif a >= 10:
        num = 2
    else:
        num = 1

print(num)

temp = a
for i in range(num):
    rest = temp // 10
    print(temp - rest * 10)
    temp = rest
```

```python
# 给定一个不超过5位的正整数，判断该数的位数，依次打印出个位、十位、百位、千位、万位的数字
# 从万位打到个位

a = int(input('>>> '))

if a >= 1000:
    if a >= 10000:
        num = 5
    else:
        num = 4
else:
    if a >= 100:
        num = 3
    elif a >= 10:
        num = 2
    else:
        num = 1

print(num)

temp = a
for i in range(num):
    print(temp // 10 ** (num - 1 - i))
    temp %= 10 ** (num - 1 - i)
```

```python
# 给定一个不超过5位的正整数，判断该数的位数，依次打印出个位、十位、百位、千位、万位的数字
# 循环的时候就给计算了
num = int(input('>>> '))
w = 10000
length = 5
flag = False

while w:
    t = num // w
    if flag:
        print(t)
    else:
        if t:
            print(t)
            flag = True
        else:
            length -= 1
    num %= w
    w //= 10
print('length:', length)
```
