```python
# 给定一个不超过5位的正整数，判断该数的位数，依次打印出个位、十位、百位、千位、万位的数字
# 法1
a = int(input('>>> '))

if a >= 1000:
    if a >= 10000:
        print(5)
    else:
        print(4)
else:
    if a >= 100:
        print(3)
    elif a >= 10:
        print(2)
    else:
        print(1)

wan = a // 10000
qian = (a - wan * 10000) // 1000
bai = (a - wan * 10000 - qian * 1000) // 100
shi = (a - wan * 10000 - qian * 1000 - bai * 100) // 10
ge = a - wan * 10000 - qian * 1000 - bai * 100 - shi * 10

print(ge, shi, bai, qian, wan)
```

```python
# 给定一个不超过5位的正整数，判断该数的位数，依次打印出个位、十位、百位、千位、万位的数字
# 循环的时候就给计算了
num = int(input('>>> '))
w = 10000
length = 5
flag = False

while w:
    t = num // w
    if flag:
        print(t)
    else:
        if t:
            print(t)
            flag = True
        else:
            length -= 1
    num %= w
    w //= 10
print('length:', length)
```

```python
# 用户输入一个数字
# 判断是几位数
# 打印每一位数字及其重复的次数
# 依次打印每一位数字，顺序个、十、百、千、万...位
# 法 1
num = input('>>> ')
length = len(num)
for i in range(1, length + 1):
    i = -i
    print('{} appears {} times.'.format(num[i], num.count(num[i])))

# 如果这个数字有 n 位，则要遍历 n 次，每次遍历为了计算出现的次数则还要从头遍历一遍，即 n 次
# 总共要计算 n ** 2 次，效率低
```

```python
# 法 2
num = input('>>> ')
length = len(num)
# 存储每个数字出现的次数
lst = []
for i in range(10):
    count = num.count(str(i))
    lst.append(count)
for j in range(1, length + 1):
    j = -j
    print('{}: appears {} times.'.format(num[j], lst[int(num[j])]))
    
# 如果这个数字有 n 位，计算 10 个数字遍历 10 次，每次要遍历整个数字，即 n 次
# 总共要计算 10 * n 次，效率比上一种方法高
```

```python
# 法 3
num = input('>>> ')
length = len(num)
# 存储每个数字出现的次数
lst = [0] * 10
for i in range(length):
    lst[int(num[i])] += 1
# for j in range(1, length + 1):
#     print('{} appeared {} times.'.format(num[-j], lst[int(num[-j])]))

# 这里的 reversed() 是惰性求值，实际上并不会生成一个新的列表，只是返回一个可迭代对象
# 所以这样使用不会占用额外空间和时间，不会降低程序运行效率
# Python 3 中建议这样使用

for j in reversed(num):
    print('{} appeared {} times.'.format(j, lst[int(j)]))
    
# 如果这个数字有 n 位，遍历每个数字的同时计算这个数字出现的次数
# 总共要计算 n 次，效率比上一种方法高
```

```python
# 输入5个数字，打印每个数字的位数，将这些数字排序打印，要求升序打印
nums_str = []
nums_int = []
for i in range(5):
    nums_str.append(input('>>> '))
for j in range(5):
    print('Number {}: {} has {} digits.'.format(j + 1, nums_str[j], len(nums_str[j])))
for k in nums_str:
    nums_int.append(int(k))
for m in range(4):
    flag = False
    for n in range(4 - m):
        if nums_int[n + 1] < nums_int[n]:
            nums_int[n + 1], nums_int[n] = nums_int[n], nums_int[n + 1]
            flag = True
    if not flag:
        break
for x in nums_int:
    print(x)
```
